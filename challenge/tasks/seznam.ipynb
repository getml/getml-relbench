### Dataset Description
> <span style="font-weight: 500; color: #3b3b3b;">ⓘ️&nbsp; Generated by `gpt-4o`</span>
>
> The *Seznam* dataset represents online advertisement expenditures from Seznam.cz, a web portal and search engine in the Czech Republic. It is structured to analyze financial transactions related to advertising.
> 
> - *Data Model*: The dataset consists of 4 tables: `dobito`, `probehnuto`, `probehnuto_mimo_penezenku`, and `client`. These tables capture details about prepaid amounts, charges from the wallet, charges outside the wallet, and client information.
> 
> - *Task*: The primary task associated with this dataset is *regression*, with the target column being `kc_proklikano` in the `probehnuto` table.
> 
> - *Column Types*: The dataset includes various data types:
>   - *Numeric*: `int`, `decimal`
>   - *String*: `varchar`
>   - *Temporal*: `date`
> 
> - *Metadata*:
>   - Size: 146.8 MB
>   - Number of Tables: 4
>   - Number of Rows: 2,681,983
>   - Number of Columns: 14
>   - Missing Values: Yes
>   - Compound Keys: No
>   - Loops: No
>   - Instance Count: 1,458,233
>   - Target Table: `probehnuto`
>   - Target Column: `kc_proklikano`
>   - Target ID: `client_id, sluzba`
>   - Target Timestamp: `month_year_datum_transakce`
> 
> This dataset is publicly available and can be accessed through a MariaDB client. It is used for analyzing advertisement spending and predicting transaction amounts.
