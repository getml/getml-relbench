### Dataset Description

>
> *Data Model:*
> 
> The *Dallas* dataset consists of three tables: `officers`, `subjects`, and `incidents`. These tables provide detailed information about officer-involved shootings as disclosed by the Dallas Police Department.
> 
> - **officers**: Contains columns `case_number` (varchar), `race` (char), `gender` (char), `last_name` (varchar), `first_name` (varchar), and `full_name` (varchar). This table provides information about the officers involved in incidents.
> 
> - **subjects**: Includes `case_number` (varchar), `race` (char), `gender` (char), `last_name` (varchar), `first_name` (varchar), and `full_name` (varchar). It details the subjects or suspects involved in the incidents.
> 
> - **incidents**: Contains `case_number` (varchar), `date` (date), `location` (varchar), `subject_statuses` (varchar), `subject_weapon` (varchar), `subjects` (varchar), `subject_count` (int), `officers` (varchar), `officer_count` (int), `grand_jury_disposition` (varchar), `attorney_general_forms_url` (varchar), `summary_url` (varchar), `summary_text` (varchar), `latitude` (double), and `longitude` (double). This table provides comprehensive details about each incident.
> 
> *Task and Target Column:*
> 
> The primary task is *classification*, with the target column being `subject_statuses` in the `incidents` table. The goal is to classify the status of subjects involved in incidents.
> 
> *Column Types:*
> 
> - Varchar: `case_number`, `location`, `subject_statuses`, `subject_weapon`, `subjects`, `officers`, `grand_jury_disposition`, `attorney_general_forms_url`, `summary_url`, `summary_text`
> - Char: `race`, `gender`
> - Integer: `subject_count`, `officer_count`
