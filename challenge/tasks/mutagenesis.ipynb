# Task: mutagenesis
### Dataset Description
> <span style="font-weight: 500; color: #3b3b3b;">ⓘ️&nbsp; Generated by `gpt-4o`</span>
>
> *Data Model:*
> 
> The *mutagenesis* dataset consists of three tables: `bond`, `atom`, and `molecule`. These tables provide information about molecular structures and their mutagenic properties.
> 
> - **bond**: Contains `atom1_id` (char), `atom2_id` (char), and `type` (int). This table details the bonds between atoms in a molecule.
> 
> - **atom**: Includes `atom_id` (char), `molecule_id` (char), `element` (char), `type` (int), and `charge` (double). It provides information about individual atoms within molecules.
> 
> - **molecule**: Contains `molecule_id` (char), `ind1` (int), `inda` (int), `logp` (float), `lumo` (double), and `mutagenic` (char). This table classifies molecules based on their mutagenic properties.
> 
> *Task and Target Column:*
> 
> The primary task is *classification*, with the target column being `mutagenic` in the `molecule` table. The goal is to classify molecules as mutagenic or non-mutagenic.
> 
> *Column Types:*
> 
> - Char: `atom1_id`, `atom2_id`, `atom_id`, `molecule_id`, `element`, `mutagenic`
> - Integer: `type`, `ind1`, `inda`
> - Float: `logp`
> - Double: `charge`, `lumo`
> 
> *Metadata:*
> 
> - **Size**: 900 KB
> - **Number of Tables**: 3
> - **Number of Rows**: 10,324
> - **Number of Columns**: 14
> - **Missing Values**: No
> - **Instance Count**: 188
> - **Target Table**: `molecule`
> - **Target Column**: `mutagenic`
> 
> This dataset is used in the medicine domain to analyze and classify chemical compounds based on their potential mutagenic effects.
