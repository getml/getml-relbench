> The *financial* dataset consists of eight tables: `loan`, `order`, `trans`, `disp`, `card`, `account`, `client`, and `district`. These tables provide comprehensive information about financial transactions, loans, and client details.
> 
> - **loan**: Contains `loan_id` (int), `account_id` (int), `date` (date), `amount` (int), `duration` (int), `payments` (decimal), and `status` (varchar). This table details loan information and outcomes.
> 
> - **order**: Includes `order_id` (int), `account_id` (int), `bank_to` (varchar), `account_to` (int), `amount` (decimal), and `k_symbol` (varchar). It records orders related to accounts.
> 
> - **trans**: Contains `trans_id` (int), `account_id` (int), `date` (date), `type` (varchar), `operation` (varchar), `amount` (int), `balance` (int), `k_symbol` (varchar), `bank` (varchar), and `account` (int). This table logs transactions.
> 
> - **disp**: Includes `disp_id` (int), `client_id` (int), `account_id` (int), and `type` (varchar). It links clients to accounts.
> 
> - **card**: Contains `card_id` (int), `disp_id` (int), `type` (varchar), and `issued` (date). This table provides card details.
> 
> - **account**: Includes `account_id` (int), `district_id` (int), `frequency` (varchar), and `date` (date). It details account information.
> 
> - **client**: Contains `client_id` (int), `gender` (varchar), `birth_date` (date), and `district_id` (int). This table provides client demographics.
> 
> - **district**: Includes `district_id` (int) and various demographic and economic indicators (varchar, int, decimal).
