Now, we can inspect all tables and annotate the columns with [roles](https://getml.com/latest/user_guide/concepts/annotating_data/).

The population table (`molecule`). We already set the `target` role for the target (`class`). If the task is a multiclass classification,
we split the target column into multiple columns in an one-vs-all fashion. In this case, the original target is still avaiable as `class`.

```python
molecule_pdf = molecule_raw.to_pandas()
molecule_pdf["split"] = molecule_pdf["split"].replace("val", "test")
molecule_pdf["molecule_name_2"] = molecule_pdf["molecule_name"].copy()
molecule = getml.data.DataFrame.from_pandas(molecule_pdf, name="molecule")
molecule
```

```python
molecule.set_role(["class"], getml.data.roles.target)
molecule.set_role(["molecule_name"], getml.data.roles.join_key)
molecule.set_role(["molecule_name_2"], getml.data.roles.text)
molecule
```

The next step is to define the data model. Refer to [https://relational.fel.cvut.cz/dataset/MuskSmall](https://relational.fel.cvut.cz/dataset/MuskSmall)
for a description of the dataset.

```python
dm = getml.data.DataModel(population=molecule.to_placeholder())
dm.add(getml.data.to_placeholder(conformation))

dm.population.join(dm.conformation, on="molecule_name")
dm
```

Now we can create the container and add the tables to it.

```python
container = getml.data.Container(population=molecule, split=molecule.split)
container.add(conformation)

container
```

```python
fast_prop = getml.feature_learning.FastProp(
    loss_function=getml.feature_learning.loss_functions.CrossEntropyLoss,
)

predictor = getml.predictors.XGBoostClassifier()
```

```python
text_field_splitter = getml.preprocessors.TextFieldSplitter()

pipe = getml.Pipeline(
    tags=['add-moleculate-name-2-as-text'],
    data_model=dm,
    feature_learners=[fast_prop],
    preprocessors=[text_field_splitter],
    predictors=[predictor],
    include_categorical=True,
)

pipe
```

```python
# getml.pipeline.delete('ZcUDIh')
```

```python
pipe.fit(container.train)
```

```python
pipe.score(container.test)
```
