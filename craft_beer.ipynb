To load the dataset, we use the `load_ctu_dataset` function from the `utils`
module. This function returns a tuple with the population table as the first
element and the a dictionary of peripheral tables as the second element.

```python
breweries_raw, peripheral = load_ctu_dataset("CraftBeer")

(
    beers,
) = peripheral.values()
```

Now, we can inspect all tables and annotate the columns with [roles](https://getml.com/latest/user_guide/concepts/annotating_data/).

The population table (`breweries`). We already set the `target` role for the target (`state`). If the task is a multiclass classification,
we split the target column into multiple columns in an one-vs-all fashion. In this case, the original target is still avaiable as `state`.

```python
breweries_pdf = breweries_raw.to_pandas()
breweries_pdf["split"] = breweries_pdf["split"].replace("val", "test")
breweries = getml.data.DataFrame.from_pandas(breweries_pdf, name="breweries")
breweries
```

```python
# TODO: Annotate remaining columns with roles
breweries.set_role(["city"], getml.data.roles.categorical)
breweries.set_role(["id"], getml.data.roles.join_key)
breweries.set_role(["name"], getml.data.roles.text)
breweries.set_role(["state=0", "state=1", "state=2", "state=3", "state=4", 
                    "state=5", "state=6", "state=7", "state=8", "state=9", 
                    "state=10", "state=11", "state=12", "state=13", "state=14", 
                    "state=15", "state=16", "state=17", "state=18", "state=19", 
                    "state=20", "state=21", "state=22", "state=23", "state=24", 
                    "state=25", "state=26", "state=27", "state=28", "state=29", 
                    "state=30", "state=31", "state=32", "state=33", "state=34", 
                    "state=35", "state=36", "state=37", "state=38", "state=39", 
                    "state=40", "state=41", "state=42", "state=43", "state=44", 
                    "state=45", "state=46", "state=47", "state=48", "state=49", 
                    "state=50"], getml.data.roles.target)

# breweries.set_role(["state=0"], getml.data.roles.target)
breweries
```

Peripheral tables,

```python
# TODO: Annotate columns with roles
beers.set_role(["brewery_id"], getml.data.roles.join_key)
beers.set_role(["abv", "ibu", "ounces"], getml.data.roles.numerical)
beers.set_role(["style"], getml.data.roles.categorical)
beers.set_role(["name"], getml.data.roles.text)

beers
```

The next step is to define the data model. Refer to [https://relational.fel.cvut.cz/dataset/CraftBeer](https://relational.fel.cvut.cz/dataset/CraftBeer)
for a description of the dataset.

Now we can create the container and add the tables to it.

```python
container = getml.data.Container(population=breweries, split=breweries.split)
container.add(beers=beers)
container.freeze()
container
```

```python
dm = getml.data.DataModel(population=breweries.to_placeholder("population"))
dm.add(getml.data.to_placeholder(beers=beers))

# TODO
dm.population.join(dm.beers, on=("id", "brewery_id"))

dm
```

```python
fast_prop = getml.feature_learning.FastProp(
    loss_function=getml.feature_learning.loss_functions.CrossEntropyLoss,
)

predictor = getml.predictors.XGBoostClassifier()
```

```python
pipe = getml.Pipeline(
    tags=['all-targets', 'rerun-to-repeat-score-of-0.6-without-pipleline.load(previous)'],
    data_model=dm,
    feature_learners=[fast_prop],
    predictors=[predictor]
    # include_categorical=True,
)

pipe
```

```python
pipe.fit(container.train)
```

```python
def prob_to_acc(prob, actual_labels, class_label) -> float:
    ix_max = np.argmax(prob, axis=1)
    predicted_labels = np.asarray([class_label[ix] for ix in ix_max])
    return (actual_labels == predicted_labels).sum() / len(actual_labels)
```

```python
actual_labels_test = breweries[breweries["split"] == "test"].state.to_numpy()
class_label = breweries.state.unique()
test_acc = prob_to_acc(predictions_prob, actual_labels_test, class_label)
test_acc
```
